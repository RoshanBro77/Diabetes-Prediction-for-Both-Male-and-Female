# Functions to convert input into numeric value and validate if age is +integer or not

```python

def isNumber(ans):
    
    if ans.strip().isdigit():
        
        return ans
    else:
        
        return 3


def convertGender(ans):
    if(ans == "Male" or ans == "male" or ans == "m" or ans == "M"):
        print(ans)
        
        return 1
    elif(ans == "Female" or ans == "female" or ans == "f" or ans == "F"):
        print(ans)
        
        return 0
    else:
        valid=False
        return 3


def convert(ans):
    if(ans == "Yes" or ans == "y" or ans == "Y" or ans == "yes"):
        print(ans)
       
        return 1

    elif(ans == "No" or ans == "n" or ans == "N" or ans == "no"):
        print(ans)
      
        return 0
    else:
        
        return 3

    
```
